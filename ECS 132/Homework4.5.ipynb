```r
# Load the Old Faithful dataset
data <- faithful

# 1. Plot the frequency histogram of the eruption duration with breaks of 1
hist(data$eruptions, breaks = seq(min(data$eruptions), max(data$eruptions) + 1, by = 1),
     xlab = "Eruption Duration (mins)", ylab = "Frequency", main = "Frequency Histogram")

# 2. Draw the boxplot of the eruption duration
boxplot(data$eruptions, main = "Boxplot of Eruption Duration", ylab = "Eruption Duration (mins)",
        whisklty = 1, whisklwd = 1, range = 0.2)

# 3. Calculate the values for the 95, 97, and 99 quantiles of the eruption duration
quantiles <- quantile(data$eruptions, probs = c(0.95, 0.97, 0.99))
print(quantiles)

# 4. Scatter plot of current eruption duration vs. duration of the next eruption
x <- data$eruptions[1:(length(data$eruptions) - 1)]
y <- data$eruptions[2:length(data$eruptions)]
plot(x, y, xlab = "Current Eruption Duration (mins)", ylab = "Next Eruption Duration (mins)",
     main = "Scatter Plot of Current vs. Next Eruption Duration")
abline(h = 3, col = "red", lty = 2) # Horizontal line at y=3 mins
abline(v = 3, col = "blue", lty = 2) # Vertical line at x=3 mins

# 5. Calculate the probabilities
long_long <- sum(data$eruptions[1:(length(data$eruptions) - 1)] > 3 & data$eruptions[2:length(data$eruptions)] > 3) / sum(data$eruptions > 3)
long_short <- sum(data$eruptions[1:(length(data$eruptions) - 1)] > 3 & data$eruptions[2:length(data$eruptions)] <= 3) / sum(data$eruptions > 3)
short_long <- sum(data$eruptions[1:(length(data$eruptions) - 1)] <= 3 & data$eruptions[2:length(data$eruptions)] > 3) / sum(data$eruptions <= 3)
short_short <- sum(data$eruptions[1:(length(data$eruptions) - 1)] <= 3 & data$eruptions[2:length(data$eruptions)] <= 3) / sum(data$eruptions <= 3)

print(paste("Probability of a long eruption followed by a long eruption:", long_long))
print(paste("Probability of a long eruption followed by a short eruption:", long_short))
print(paste("Probability of a short eruption followed by a long eruption:", short_long))
print(paste("Probability of a short eruption followed by a short eruption:", short_short))
```
