```r
# Function to simulate hat matching problem
simulate_hat_matching <- function(n, num_experiments) {
  count_matches <- 0
  for (i in 1:num_experiments) {
    hats <- sample(1:n, n, replace = FALSE)
    if (sum(hats == 1:n) > 0) {
      count_matches <- count_matches + 1
    }
  }
  return(count_matches / num_experiments)
}

# Set parameters
num_experiments <- 1000
n_values <- seq(1, 100, by = 1)

# Perform simulations
probabilities <- sapply(n_values, function(n) simulate_hat_matching(n, num_experiments))

# Calculate asymptotic value
asymptotic_value <- 1 - 1/exp(1)

# Plot results
plot(n_values, probabilities, type = "l", xlab = "Number of people (n)", ylab = "Probability of at least one match", main = "Probability of Hat Matching")
abline(h = asymptotic_value, col = "red")
legend("topright", legend = c("Simulation", "Asymptotic Value"), col = c("black", "red"), lty = 1)
```
