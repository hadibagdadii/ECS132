```r
n <- 10  # number of trials
p <- 0.4  # probability of success
x <- 0:n  # possible values of X

pmf <- dbinom(x, size = n, prob = p)
plot(x, pmf, type = "h", main = "Probability Mass Function of X", xlab = "Number of Survivors", ylab = "Probability")
```

```r
cdf <- pbinom(x, size = n, prob = p)
plot(x, cdf, type = "s", main = "Cumulative Distribution Function of X", xlab = "Number of Survivors", ylab = "Cumulative Probability")
```

```r
prob_at_least_8 <- 1 - pbinom(7, size = n, prob = p)
print(prob_at_least_8)
```

```r
prob_3_to_8 <- sum(dbinom(3:8, size = n, prob = p))
print(prob_3_to_8)
```
